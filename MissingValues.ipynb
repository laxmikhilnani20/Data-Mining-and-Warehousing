AIM : To treat missing values with different techniques in python

```python
import pandas as pd
import numpy as np
print('Hello')
```

```python
dataset = pd.read_excel('dataFiles/Employee_Data.xls')
```

```python
#Cell 2
dataset.isnull().sum()
```

```python
dataset
```

```python
x=dataset.iloc[:,3:6].values
x
```

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values = 'NaN', strategy = 'median')
imputer = SimpleImputer().fit(x[:,1:3])
x[:,1:3] = imputer.transform(x[:,1:3])
print(x)
```

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values = 'NaN', strategy='most_frequent')
imputer = SimpleImputer().fit(x[:,1:3])
x[:,1:3] = imputer.transform(x[:,1:3])
print(x)
```
